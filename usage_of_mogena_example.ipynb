```python
from mogena.Mogena import MoGenA
import math
```

```python
def griewank_(xx):
    d    = len(xx)
    sum  = 0
    prod = 1
    for ii in range(d):
        xi   = xx[ii]
        sum  = sum + xi**2/4000
        prod = prod * math.cos(xi/math.sqrt(ii+1))
    out = sum - prod + 1
    return out
```

```python
def egg_(xx):
    x1    = xx[0]
    x2    = xx[1]
    term1 = -(x2+47) * math.sin(math.sqrt(math.fabs(x2+x1/2+47)))
    term2 = -x1 * math.sin(math.sqrt(math.fabs(x1-(x2+47))))
    out   = term1 + term2
    return out
```

```python
def usage():
    ga = MoGenA(fitness_func=egg_, number_of_generation=250, number_of_variables=2,
                population_size=500,
                lower_boundaries=[-512,-512],
                upper_boundaries=[512,512],use_saved_data=True)
    return {'best_fitness_ever':ga.best_fitness_ever,
            'best_individual_decoded_ever':ga.best_individual_decoded_ever}
```

```python
usage()
```
